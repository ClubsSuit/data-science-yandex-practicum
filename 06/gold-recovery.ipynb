### –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç–∏ —Ä–∞—Å—á–µ—Ç–∞ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∏—è ```recovery```

–†–∞—Å—á–µ—Ç **MAE** –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ ```rougher.output.recovery```

````python
def recovery(c, f, t):
    return (c*(f-t))/((f)*(c-t)) * 100
````

````python
rougher_recovery_calc = recovery(data_full['rougher.output.concentrate_au'], 
                                               data_full['rougher.input.feed_au'], 
                                               data_full['rougher.output.tail_au']
                                              )

display(mean_absolute_error(data_full['rougher.output.recovery'], rougher_recovery_calc))
````

–†–∞—Å—á–µ—Ç **MAE** –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ ```final.output.recovery```

````python
final_recovery_calc = recovery(data_full['final.output.concentrate_au'], 
                                               data_full['rougher.input.feed_au'], 
                                               data_full['final.output.tail_au']
                                              )
````

````python
display(mean_absolute_error(data_full['final.output.recovery'], final_recovery_calc))
````

## –ú–æ–¥–µ–ª—å

### –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ —Ñ–∏–Ω–∞–ª—å–Ω–æ–π –º–µ—Ç—Ä–∏–∫–∏ sMAPE

````python
def smape(y_true,y_pred):
    return np.mean(abs(y_true - y_pred) / ((abs(y_true) + abs(y_pred)) / 2)) * 100
````

````python
def final_smape(r, f):
    return (0.25*r + 0.75*f)
````

````python
custom_score = make_scorer(smape, greater_is_better=False)
````
